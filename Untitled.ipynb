```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('raw.xlsx', sheet_name='processed')
```

```python
df
```

```python
conditions = [
    (df.loc[(df['stop_loss'] <= df['min']) & (df['sell'] <= df['max'])]),
    (df.loc[(df['stop_loss'] >= df['min'])])
]
```

```python
df['profit'] = 0
```

```python
for index, row in df.iterrows():
    if (row['stop_loss'] <= row['min']) and (row['sell'] <= row['max']):
        df.loc[index, 'profit'] = 100 * (row['sell'] - row['buy']) / row['buy']
    elif (row['stop_loss'] >= row['min']):
        df.loc[index, 'profit'] = 100 * (row['stop_loss'] - row['buy']) / row['buy']
```
